```python
import re
# Open the file in read mode ('r')
with open('/content/drive/My Drive/Colab Notebooks/text.txt', 'r', encoding="utf-8") as file:
    content = file.read()
text_lines = content.split(sep='\n') # Get the text lines into a list
result = [] # Init the result list
for s in text_lines:
    ending = re.search(r'\W$', s).group() # Check if the current sentence ends with a punctuation mark == not a letter, \W
    if ending:
        result.append(' '.join(s[:-1].split(' ')[::-1]) + ending) # If not a letter, split the sentence without the last symbol by spaces, reverse, join with space delimeters, add the punctuation and append to the result
    else:
        result.append(' '.join(s.split(' ')[::-1])) # If a letter, same for the whole sentence
print('Исходный текст:', content, 'Результат:', '\n'.join(result), sep = '\n\n') # Print the source and the result




```

**Сжатие строки**

```python
from itertools import groupby

def str_compress(src_string):
    result = '' # Init the result list
    if src_string:
        for char, group in groupby(src_string): # Create an iterator over the string like (character, [character, character, ...])
            count = len(list(group)) # How long is the group?
            result += f'{count}{char}' if count > 1 else char # Append the f-string of the count and the char if the char repeats, and the char if not
        return result # If the input string is not empty, return the compressed string
    else:
        print ('Введена пустая строка') # Return a message and an empty string to avoid "None"
        return ''
print (str_compress('A9!@###  pqR77$$$%% xyz&&&***--==++@@!!??'))
print (str_compress('abcd'))
print (str_compress('a'))
print (str_compress(''))
```

**Проверка скобок**

```python
def correct_brackets(src_string):
    brackets = {')': '(', ']': '[', '}': '{'} # Dictionary with closing: opening brackets
    br_only = [char for char in src_string if char in '()[]{}'] # Remove everything except the brackets from the string
    stack = [] # Init the stack
    for char in br_only:
        if char in brackets.values(): # Is it an opening bracket?
            stack.append(char) # Add the bracket to the stack
        elif stack[-1] != brackets[char]:
            return False # If the last bracket in the stack is not the pairing closing bracket => False
        else:
            stack.pop() # If the closing bracket matches the opening one already in the stack, pop it and move on
    return True # If no errors => True, otherwise the prior return will return False at some point
print(correct_brackets('''((2 + 3) * [5 / {7 - (1)}])'''))
print(correct_brackets('''[{()}([]{})]'''))
print(correct_brackets('''function(x, y] { return x+y; }'''))
print(correct_brackets('''[(])'''))
```

**Генератор паролей**

```python
import secrets
import string
def password_generator(pass_len):
    try:
        num = int(pass_len)
    except:
        print('Введите целое число в качестве длины пароля')
        return ''
    if num > 0:
        c_set = string.printable[:-5] # All printable characters that can be entered from the keyboard
        result = [secrets.choice(c_set)] # Add the first symbol
        while len(result) < num: # Add symbols until the set length is reached
            next_set = ''.join([s for s in c_set if s != result[-1]]) # Get the next char that is guaranteed not to match the previous one
            result += secrets.choice(next_set) # Append the next char to the resulting string
        return ''.join(result)
    else:
        print('Длина пароля должна быть больше 0')
        return ''
print(password_generator(pass_len = input('Введите длину пароля для генерации: ')))
```

**Почти палиндром**

```python
def is_palindrome(src_string):      # Check if the string is a palindrome as is
    ch_string = ''.join([s for s in src_string if s != ' '])    # Remove spaces from the string
    s_len = len(ch_string)
    # Take the first half of the string and compare with the 2nd part taken in reverse. s_len // 2 to accont for odd and even length strings.
    # If the length is odd, s_len % 2 == 1 and the central symbol is omitted, because it doesn't decide if the string is a palindrome.
    return ch_string[:s_len // 2] == ch_string[s_len // 2 + s_len % 2:][::-1]
def is_part_palindrome(ch_string):  # Check if the string is a palindrome as is or if any letter is removed
    print(ch_string)
    ch_string = ch_string.lower() # Even though the case doesn't match if the string is read in reverse, the string is still perceived as a palindrome
    if is_palindrome(ch_string):    # Palindrome as is?
        print('Строка является палиндромом')
    else:
        fl = False
        for i in range(len(ch_string)-1):   # Try to remove every character one at a time except for the last character
            letter_removed = ch_string[:i] + ch_string[i+1:] # Remove i-th character
            if is_palindrome(letter_removed):   # Becomes a palindrome?
                print(f'Строка является палиндромом при удалении {i + 1}-го символа - {ch_string[i]}')
                fl = True
        if is_palindrome(ch_string[:-1]):   # Try to remove the last character and check for palindrome
            fl = True
            print(f'Строка является палиндромом при удалении {len(ch_string)}-го символа - {ch_string[-1]}')
        if not fl:      # None of the previous makes the string a palindrome
            print('Строка не является палиндромом')
is_part_palindrome('Лёша на полке кклопа нашёл')
is_part_palindrome('Тонет енот')
is_part_palindrome('Строка не палиндром')
```

**Задача Шредингера**

```python
import re
import random
threshold = -1      # Init for user input validation loop
while threshold <= 0:   # Until a float is input
    threshold = input('Введите процент сокращения текста в процентах: ')
    try:
        threshold = float(threshold)    # Is the input a float?
    except:     # Not a number entered
        print('Введите положительное число')
        threshold = -1      # If not reassigned -1, the loop will break
        continue            # Avoid the following print, because it's not valid here
    if threshold <= 0:      # A negative entered?
        print('Введите положительное число')
# Open the file in read mode ('r')
with open('/content/drive/My Drive/Colab Notebooks/text.txt', 'r', encoding="utf-8") as file:
    content = file.read()
l_con = len(content)    # How long is the text?
reduction = 0           # Init the reduction measure
cur_content = content   # Init the current text
while reduction < threshold:
    # Make a list of start positions of every word in the text. (?<![\w-])[\w-]+(?![\w-]) == any number of letters between the
    # word boundaries == matches the whole word
    word_pos = [i.start() for i in re.finditer(r'(?<![\w-])[\w-]+(?![\w-])', cur_content)]
    if not word_pos:    # If at some point all words were replaced and the entered reduction is not achieved == maximum possible reduction reached
        print('Заданный процент сокращения обеспечить невозможно')
        break
    n = random.choice(word_pos)     # Choose a random word start to make a replacement
    # Replacing the word by slicing the string at the start position and merging the result. Replace only one word, count=1
    new_content = cur_content[:n] + re.sub(r'(?<![\w-])[\w-]+(?![\w-])', '...', cur_content[n:], count=1)
    reduction = (l_con - len(new_content)) / l_con * 100    # Update the achieved reduction against the initial text
    cur_content = new_content       # Update the text to be used for the next cycle iteration
with open('/content/drive/My Drive/Colab Notebooks/text_reduced.txt', 'w', encoding="utf-8") as file:
    file.write(new_content)     # Write the resulting text into a new file
print('\nИсходный текст:', content, 'Результат (записан в файл text_reduced.txt):', new_content, f'Процент сокращения: {reduction}%', sep = '\n\n') # Print the source and the result





```
