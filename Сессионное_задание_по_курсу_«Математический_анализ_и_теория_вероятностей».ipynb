Найти условные экстремумы функции
$$
z(x,y) = e^{-x - y},
$$
если задано условие
$$
\frac{x^2}{4} + y^2 = 5, \quad g(x,y) = \frac{x^2}{4} + y^2 - 5.
$$

Составим функцию Лагранжа:
$$
L(x,y,\lambda) = z(x,y) + \lambda g(x,y),\\
L(x,y,\lambda) = e^{-x - y} + \lambda\left(\frac{x^2}{4} + y^2 - 5\right).
$$

Для стационарных точек выполняется система:
$$
\begin{cases}
\frac{\partial L}{\partial x} = 0, \\
\frac{\partial L}{\partial y} = 0, \\
\frac{\partial L}{\partial \lambda} = 0.
\end{cases}
$$

Частные производные:
$$
\frac{\partial L}{\partial x} = -e^{-x - y} + \lambda\frac{x}{2}, \qquad
\frac{\partial L}{\partial y} = -e^{-x - y} + 2\lambda y, \qquad
\frac{\partial L}{\partial \lambda} = \frac{x^2}{4} + y^2 - 5.
$$

Получаем систему:
$$
\begin{cases}
-e^{-x - y} + \lambda\dfrac{x}{2} = 0, \\
-e^{-x - y} + 2\lambda y = 0, \\
\dfrac{x^2}{4} + y^2 = 5.
\end{cases}
$$

Из первых двух уравнений:
$$
\lambda = \frac{2e^{-x - y}}{x}, \qquad \lambda = \frac{e^{-x - y}}{2y}.
$$

Приравниваем и сокращаем на $e^{-x-y} \neq 0$:
$$
\frac{2}{x} = \frac{1}{2y} \Rightarrow 4y = x.
$$

Подставляем в третье уравнение:
$$
\frac{(4y)^2}{4} + y^2 = 5 \Rightarrow 4y^2 + y^2 = 5 \Rightarrow 5y^2 = 5 \Rightarrow y^2 = 1.
$$

Следовательно:
$$
y = \pm 1, \qquad x = 4y.
$$

Стационарные точки:
$$
(4,1), \qquad (-4,-1).
$$

Множители Лагранжа из $\frac{\partial L}{\partial x} = 0$:
$$
\lambda = \frac{2e^{-x - y}}{x}.
$$

Для $(4,1)$:
$$
\lambda = \frac{2e^{-5}}{4} = \frac{1}{2}e^{-5}.
$$

Для $(-4,-1)$:
$$
\lambda = \frac{2e^{5}}{-4} = -\frac{1}{2}e^{5}.
$$

Вторые производные функции Лагранжа:
$$
\frac{\partial^2 L}{\partial x^2} = e^{-x - y} + \frac{\lambda}{2}, \qquad
\frac{\partial^2 L}{\partial y^2} = e^{-x - y} + 2\lambda, \qquad
\frac{\partial^2 L}{\partial x \partial y} = e^{-x - y}.
$$

Условие связи:
$$
dg = \frac{\partial g}{\partial x}dx + \frac{\partial g}{\partial y}dy \Rightarrow
dg = \frac{x}{2}\,dx + 2y\,dy = 0.
$$
Отсюда
$$
dy = -\frac{x}{4y}dx.
$$

Второй дифференциал функции Лагранжа:
$$
d^2L = \frac{\partial^2 L}{\partial x^2}(dx)^2 + 2\frac{\partial^2 L}{\partial x \partial y}dx\,dy + \frac{\partial^2 L}{\partial y^2}(dy)^2.
$$

Подставляем связь $dy = -\dfrac{x}{4y}dx$:
$$
d^2L = \left(\frac{\partial^2 L}{\partial x^2} - 2\frac{x}{4y}\frac{\partial^2 L}{\partial x \partial y} + \frac{x^2}{16y^2}\frac{\partial^2 L}{\partial y^2}\right)(dx)^2.
$$

Точка $(4,1)$, $\lambda = \frac{1}{2}e^{-5}$:
$$
\frac{\partial^2 L}{\partial x^2} = e^{-5} + \frac{1}{4}e^{-5} = \frac{5}{4}e^{-5}, \quad
\frac{\partial^2 L}{\partial y^2} = e^{-5} + e^{-5} = 2e^{-5}, \quad
\frac{\partial^2 L}{\partial x \partial y} = e^{-5}.
$$

Подставляем $x=4$, $y=1$:
$$
d^2L = \left(\frac{5}{4}e^{-5} - 2\cdot\frac{4}{4\cdot 1}e^{-5} + \frac{16}{16\cdot 1^2}2e^{-5}\right)(dx)^2
= \frac{5}{4}e^{-5}(dx)^2 > 0.
$$
Следовательно, $(4,1)$ — точка условного минимума.

Точка $(-4,-1)$, $\lambda = -\frac{1}{2}e^{5}$:
$$
\frac{\partial^2 L}{\partial x^2} = e^{5} - \frac{1}{4}e^{5} = \frac{3}{4}e^{5}, \quad
\frac{\partial^2 L}{\partial y^2} = e^{5} - e^{5} = 0, \quad
\frac{\partial^2 L}{\partial x \partial y} = e^{5}.
$$

Подставляем $x=-4$, $y=-1$:
$$
d^2L = \left(\frac{3}{4}e^{5} - 2\cdot\frac{-4}{4\cdot (-1)}e^{5} + \frac{16}{16\cdot (-1)^2}\cdot 0\right)(dx)^2
= -\frac{5}{4}e^{5}(dx)^2 < 0.
$$
Следовательно, $(-4,-1)$ — точка условного максимума.

Ответ:
$$
z(4,1) \approx 0,0067 \text{ — условный минимум}, \qquad z(-4,-1) \approx 148,4132 \text{ — условный максимум}.
$$

```python
func = lambda x: x**4 + 3*x**3 - 12*x**2 + 7*x - 2
dfunc = lambda x: 4*x**3 + 9*x**2 - 24*x + 7
ddfunc = lambda x: 12*x**2 + 18*x - 24
x0 = 10
count = 0
x_current = x0
eps = 0.0001
f = dfunc(x_current)

while abs(f) > eps:
    x_current = x_current - f / ddfunc(x_current)
    f = dfunc(x_current)
    count += 1
    print (x_current)
print ('Точка локального минимума:', x_current)
print ('Первая производная в точке:', dfunc(x_current))
print ('Вторая производная в точке:', ddfunc(x_current))
print ('Значение функции в точке:', func(x_current))
```

```python
from scipy.optimize import newton
func = lambda x: x**4 + 3*x**3 - 12*x**2 + 7*x - 2
dfunc = lambda x: 4*x**3 + 9*x**2 - 24*x + 7
ddfunc = lambda x: 12*x**2 + 18*x - 24
x_min = newton(dfunc, x0 = 10, fprime = ddfunc, tol = 0.0001)
print ('Точка локального минимума: ', x_min)
print ('Локальный минимум: ', func(x_min))
```

```python
import numpy as np
X = np.array(([0, 1, 2, 3, 4, 5, 6, 7],
             [0.1, 0.3, 0.1, 0.25, 0.05, 0.05, 0.1, 0]))

# Вычисление пропущенной вероятности
X[1][7] = 1 - X[1].sum()

# Математическое ожидание - поэлементное произведение строк и сумма результатов
M = np.sum(X[0]*X[1])

# Дисперсия - среднее квадрата величины минус квадрат среднего
D = np.sum(X[0]**2*X[1]) - M*M

print (f"Пропущенная вероятность для 7 писем: {X[1][7]:.2f}")
print (f"Математическое ожидание: {M:.2f}")
print (f"Дисперсия: {D:.2f}")
```

Случайная величина $X$ — число звонков за час, описывается распределением Пуассона

В среднем 16 звонков за 8 часов, значит за 1 час
$$
\lambda = \frac{16}{8} = 2.
$$

Найти
$$
P(2 \le X \le 4) = P(X = 2) + P(X = 3) + P(X = 4).
$$

По формуле вероятности, что произойдет ровно $k$ событий, для распределения Пуассона:
$$
P(X = k) = \frac{\lambda^k}{k!} e^{-\lambda}.
$$

Подставим $\lambda = 2$:
$$
\begin{aligned}
P(X = 2) &= \frac{2^2}{2!} e^{-2} = 2 e^{-2},\\
P(X = 3) &= \frac{2^3}{3!} e^{-2} = \frac{4}{3} e^{-2},\\
P(X = 4) &= \frac{2^4}{4!} e^{-2} = \frac{2}{3} e^{-2}.
\end{aligned}
$$

Тогда
$$
P(2 \le X \le 4) = (2 + \frac{4}{3} + \frac{2}{3}) e^{-2} = 4 e^{-2} \approx 0,541.
$$

Ответ:
$$
P(2 \le X \le 4) \approx 0,541.
$$
